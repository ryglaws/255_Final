# 9. Create a loop to find routes

```python
def route(row):
    try:
        route=nx.shortest_path(G, source=row.orig_node, target=row.dest_node)
    except: 
        route=np.nan
    return route
        
ods_dist['route'] = ods_dist.apply(route, axis=1)

ods_dist
```

```python
all_ods=ods_dist.drop(columns=['Unnamed: 0'])
all_ods
```

```python
all_ods['distmi']=all_ods['dist']/1609
```
